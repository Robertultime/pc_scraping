```python
def get_last_x(dt):
    a = 0.03
    b = 1
    
    T = 10
    N = int(T/dt)
    x=0
    for i in range(N):
        dz = np.random.normal() * (dt)**(1/2)
        x += dz*b + a*dt
    return x
```

```python
fat_dt = []
fit_dt = []

for i in range(10000):
    fat_dt.append(get_last_x(1))
    fit_dt.append(get_last_x(0.01))

nfat = np.array(fat_dt)
nfit = np.array(fit_dt)

print(f'1: Mean:{nfat.mean()}, Variance:{nfat.std()}')
print(f'0.001: Mean:{nfit.mean()}, Variance:{nfit.std()}')
```

```python
#Monte Carlo simulation :

S = 100
S_list = [S]
dt = 1/365
T= 1
n = int(T/dt)
mu = 0.15
vol = 0.3
N = 500
N_list = []


for i in range(N):
    S = 100
    S_list = [S]
    for i in range(n):
        eps = np.random.normal()
        S += S*(mu*dt + vol*eps*(dt**(1/2)))
        S_list.append(S)
    N_list.append(S_list)
y = [i*dt for i in range(len(S_list))]
    
    
```

```python
y = [i*dt for i in range(len(S_list))]
plt.plot(y, S_list)
```

```python
simulations = pd.DataFrame(N_list, columns=y).T
```

```python
plt.figure(figsize=(20, 20))
sns.lineplot(simulations, legend=False)
```
